```python
from conllu import parse

def sort_sentences_by_sent_id(conllu_file, output_file):
    # Lecture du fichier CoNLL-U
    with open(conllu_file, 'r', encoding='utf-8') as f:
        data = f.read()

    # Parsing du fichier CoNLL-U
    sentences = parse(data)

    # Tri des phrases selon sent_id
    sorted_sentences = sorted(sentences, key=lambda x: tuple(map(int, x.metadata['sent_id'].replace("_split", "_").split('_'))))

    # Écrit les phrases concaténées dans un fichier de sortie
    with open(output_file, "w", encoding="utf-8") as output:
        output.writelines([sentence.serialize() for sentence in sorted_sentences])

# Chemin vers le fichier CoNLL-U
conllu_file_path = "/home/ziane212/Téléchargements/dump_Gascon_23/last/1251_Coutumes_de_Bagnieres.conllu"
output_file = "/home/ziane212/Téléchargements/dump_Gascon_23/last/1251_Coutumes_de_Bagnieres_test.conllu"

# Tri des phrases selon sent_id
sorted_sentences = sort_sentences_by_sent_id(conllu_file_path, output_file)
```

```python
from conllu import parse_incr

# Fonction pour supprimer les doublons et afficher les phrases non identiques avec le même sent_id
def supprimer_doublons_et_controler(fichier_conllu, fichier_sortie):
    with open(fichier_conllu, 'r', encoding='utf-8') as f:
        phrases_par_sent_id = {}
        
        # Lire et regrouper les phrases par sent_id
        for phrase in parse_incr(f):
            sent_id = phrase.metadata['sent_id']
            if sent_id in phrases_par_sent_id:
                phrases_par_sent_id[sent_id].append(phrase)
            else:
                phrases_par_sent_id[sent_id] = [phrase]
        
        # Suppression des doublons et contrôle des phrases non identiques
        with open(fichier_sortie, 'w', encoding='utf-8') as fichier_out:
            for sent_id, phrases in phrases_par_sent_id.items():
                unique_phrases = []
                different_phrases = []
                
                for phrase in phrases:
                    tokens_form = [token['form'] for token in phrase]
                    
                    # Vérifier si une phrase similaire existe déjà
                    if any([tokens_form == [token['form'] for token in p] for p in unique_phrases]):
                        continue  # Phrase identique, on l'ignore
                    else:
                        unique_phrases.append(phrase)
                        different_phrases.append(tokens_form)  # Ajouter pour contrôle
                    
                # Si plusieurs phrases non identiques existent avec le même sent_id, on les affiche pour contrôle
                if len(different_phrases) > 1:
                    print(f"Sent ID: {sent_id} contient des phrases non identiques :")
                    for tokens in different_phrases:
                        print(" ".join(tokens))
                    print()
                
                # Écrire les phrases uniques dans le fichier de sortie
                for phrase in unique_phrases:
                    fichier_out.write(phrase.serialize())
    
    print(f"Le fichier sans doublons a été écrit dans : {fichier_sortie}")

# Exemple d'utilisation
fichier_conllu = "/home/ziane212/Téléchargements/dump_Gascon_1460_Les_Fors_anciens_de_Bearn_21.08.24/validated/1460_Les_Fors_anciens_de_Bearn_21.08.24.conllu"
fichier_sortie = "/home/ziane212/Téléchargements/dump_Gascon_1460_Les_Fors_anciens_de_Bearn_21.08.24/validated/1460_Les_Fors_anciens_de_Bearn_21.08.24_sans_doublons.conllu"

supprimer_doublons_et_controler(fichier_conllu, fichier_sortie)


# # Exemple d'utilisation
# fichier_conllu = "/home/ziane212/Téléchargements/dump_Gascon_1460_Les_Fors_anciens_de_Bearn_21.08.24/validated/1460_Les_Fors_anciens_de_Bearn_21.08.24.conllu"
# fichier_sortie = "/home/ziane212/Téléchargements/dump_Gascon_1460_Les_Fors_anciens_de_Bearn_21.08.24/validated/1460_Les_Fors_anciens_de_Bearn_21.08.24_sans_doublons.conllu"

# supprimer_doublons_et_ecrire(fichier_conllu, fichier_sortie)

```
