```python
import conllu

path_orig = "/home/ziane212/projects/AUTOMATED/bootstrapping_SRCMF/data/fro/fro_profiterole-ud-dev.conllu"
path_trans = "/home/ziane212/projects/AUTOMATED/bootstrapping_SRCMF/data/fro/out_grs_trans/fro_profiterole-ud-dev.conllu"
path_out = "/home/ziane212/projects/AUTOMATED/conversion_AUX/out_trans/fro_profiterole-ud-dev.conllu"

with open(path_orig, 'r', encoding='utf-8') as orig_file, \
    open(path_trans, 'r', encoding='utf-8') as trans_file, \
    open(path_out, 'w', encoding='utf-8') as out_file:
    orig_sentences = list(conllu.parse(orig_file.read()))
    trans_sentences = list(conllu.parse(trans_file.read()))

    for orig_sent, trans_sent in zip(orig_sentences, trans_sentences):
        if orig_sent != trans_sent:
            trans_sent.metadata['tags']="trans"
            # out_file.writelines([sentence.serialize() + "\n" for sentence in trans_sentences])
            out_file.writelines(trans_sent.serialize())
```

```python
import conllu

l_files = ["/home/ziane212/projects/AUTOMATED/bootstrapping_SRCMF/data/fro/fro_profiterole-ud-train.conllu",
            "/home/ziane212/projects/AUTOMATED/bootstrapping_SRCMF/data/fro/fro_profiterole-ud-test.conllu",
            "/home/ziane212/projects/AUTOMATED/bootstrapping_SRCMF/data/fro/fro_profiterole-ud-dev.conllu", 
            "/home/ziane212/projects/AUTOMATED/bootstrapping_SRCMF/data/frm/frm_profiterole-ud-test.conllu"
            ]

output_file_path = "/home/ziane212/projects/AUTOMATED/conversion_AUX/profiterole_AUX_form.lex"

l_form = []
for f in l_files:
    with open(f, 'r', encoding='utf-8') as infile:
        sentences = list(conllu.parse(infile.read()))

    for sent in sentences:
        for tok in sent:
            if tok['upos']=='AUX' and tok['deprel']=='aux':
                form = str(tok)#.lower()
                l_form.append(form)
        l_form = list(sorted(set(l_form)))

with open(output_file_path, 'w', encoding='utf-8') as outfile:
    # Utiliser writelines pour écrire chaque ligne séparément
    outfile.write('form\n')
    outfile.writelines([form + "\n" for form in l_form])
```

```python
import pprint
import grew

grew.init()

g = grew.graph('''graph {
  W1 [form="le", lemma="le", cat=DET];
  W2 [form="garçon", lemma="garçon", cat=NOUN];
  W3 [form="voit", lemma="voir", cat=VERB];
  W4 [form="la", lemma="le", cat=DET];
  W5 [form="maison", lemma="maison", cat=NOUN];
  }''')

r = grew.grs('''
rule set_gender {
  pattern { N [upos=NOUN, !Gender, lemma=lex.noun] }
  commands { N.Gender = lex.Gender }
}
#BEGIN lex
noun\tGender
%-------------
garçon\tMasc
maison\tFem
maison\tMasc
#END
''')

output = grew.run(r, g, 'Iter(set_gender)')

pprint.pprint (output)
```
